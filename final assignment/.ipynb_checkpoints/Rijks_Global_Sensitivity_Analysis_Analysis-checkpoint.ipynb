```python
deaths = (outcomes['A.1_Expected Number of Deaths'] + 
                                               outcomes['A.2_Expected Number of Deaths'] +
                                               outcomes['A.3_Expected Number of Deaths'] +
                                               outcomes['A.4_Expected Number of Deaths'] +
                                               outcomes['A.5_Expected Number of Deaths']
                                              )

costs = (              outcomes['A.1 Total Costs'] + 
                                         outcomes['A.2 Total Costs'] +
                                         outcomes['A.3 Total Costs'] +
                                         outcomes['A.4 Total Costs'] +
                                         outcomes['A.5 Total Costs'] +
                                         outcomes['RfR Total Costs']
                          )
```

```python
dike_model, planning_steps = get_model_for_problem_formulation(3)
```

```python
#Create SALib problem from uncertainties
uncertainties = dike_model.uncertainties
problem = get_SALib_problem(uncertainties)
print(problem)
```

```python
#Create indices for Expected Number of Deaths
indices_deaths = sobol.analyze(problem,deaths)

#Create indices for Expected Annual Damage
indices_costs = sobol.analyze(problem,costs)
```

```python
Si_filter = {k:indices_deaths[k] for k in ['ST','ST_conf','S1','S1_conf']}
Si_df = pd.DataFrame(Si_filter, index=problem['names'])
threshold = 0.01 #Set threshold to remove very low influences
Si_df = Si_df[~(Si_df['ST'] < threshold)]

sns.set_style('white')
fig, ax = plt.subplots(1)

indices = Si_df[['S1','ST']]
err = Si_df[['S1_conf','ST_conf']]

indices.plot.bar(yerr=err.values.T,ax=ax)
fig.set_size_inches(8,6)
fig.subplots_adjust(bottom=0.3)
plt.savefig('./results/sobolsensitivitydeaths_randomsampling.jpg', bbox_inches="tight")
plt.show()
```

```python
Si_filter = {k:indices_costs[k] for k in ['ST','ST_conf','S1','S1_conf']}
Si_df = pd.DataFrame(Si_filter, index=problem['names'])
threshold = 0.01 #Set threshold to remove very low influences
Si_df = Si_df[~(Si_df['ST'] < threshold)]

sns.set_style('white')
fig, ax = plt.subplots(1)

indices = Si_df[['S1','ST']]
err = Si_df[['S1_conf','ST_conf']]

indices.plot.bar(yerr=err.values.T,ax=ax)
fig.set_size_inches(8,6)
fig.subplots_adjust(bottom=0.3)
plt.savefig('./results/sobolsensitivitycosts_randomsampling.jpg', bbox_inches="tight")
plt.show()
```
